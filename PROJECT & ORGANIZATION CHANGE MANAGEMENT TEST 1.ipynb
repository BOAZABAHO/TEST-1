```python
from sklearn.linear_model import LinearRegression
```

```python
#Using identified cases for the next 24 hours
x=df[['Cases - newly reported in last 24 hours']]
y=df['Deaths - newly reported in last 24 hours']
```

```python
lm=LinearRegression()
lm.fit(x,y)
yhat=lm.predict(x)
yhat[0:5]
```

```python
#calculation of coefficient m
lm.coef_
```

```python
#intercept c
lm.intercept_
```

```python
#calculating r-squared
lm.score(x,y)
```

y=mx+c
$$Deaths - newly reported in last 24 hours=0.02106987xCases - newly reported in last 24 hours+9.95968105413516$$


```python
#Let's import the function mean_squared_error from the module metrics
from sklearn.metrics import mean_squared_error
```

```python
#RMSE
y=lm.predict(x)
MSE = mean_squared_error(df['Deaths - newly reported in last 24 hours'], y)
print('MSE is', MSE)
```

```python
#calculate RSME
#import necessary libraries
from math import sqrt
```

```python
sqrt(mean_squared_error(df['Deaths - newly reported in last 24 hours'], y))
```

# Multiple LinearRegression

Predicting Deaths-newly reported in last 24 hours in America by looking at; Cases - newly reported in last 7 days,cases - cumulative total per 100000 population, and Cases - newly reported in last 24 hours


Model formulae;
$$ùë¶=ùëè1ùë•1+ùëè2ùë•2+b3x3+ùëê$$


```python
# The different variables here are; Cases - newly reported in last 7 days, cases - cumulative total per 100000 population and Cases,Cases - newly reported in last 24 hours
z = df[['Cases - newly reported in last 7 days', 'Cases - cumulative total per 100000 population','Cases - newly reported in last 24 hours']]

```

```python
#Fitting the linear model using the three above-mentioned variables.
lm.fit(z, df['Deaths - newly reported in last 24 hours'])
```

```python
#calculation of coefficient b1,b2,b3 
lm.coef_
```

```python
#intercept 
lm.intercept_
```

```python
#calculating r-squared
lm.score(z, df['Deaths - newly reported in last 24 hours'])
```

```python
#Calculating the Mean Squared Error (MSE)
#prediction based on our model
y_predict_multifit = lm.predict(z)
#Now we can compare the predicted results with the actual results
mean_squared_error(df['Deaths - newly reported in last 24 hours'], y_predict_multifit)
```

```python
#calculating RMSE
sqrt(mean_squared_error(df['Deaths - newly reported in last 24 hours'], y_predict_multifit))
```
